```
public class RoadMap(int rows, int cols){
    char [,] map = new char [rows,cols];

    public void DrawRectangle(int x1, int y1, int x2, int y2){
        for (int i = y1; i<=y2; i++){
            for (int j = x1; j<=x2; j++){
                if (i==y1){
                    if (j==x1){
                        map [i,j] = '┌';
                    }
                    else if (j==x2){
                        map [i,j] = '┐';
                    }
                    else{
                        map [i,j] = '─';
                    }
                }
                else if (i==y2){
                    if (j==x1){
                        map [i,j] = '└';
                    }
                    else if (j==x2){
                        map [i,j] = '┘';
                    }
                    else{
                        map [i,j] = '─';
                    }
                }
                else if (j==x1 || j==x2){
                    map [i,j] = '│';
                }
            }
        }
        Redraw(x1, y1, x2, y2);
    }

    public void DrawLineGorizontal(int x1, int x2, int y){
        for (int i = x1; i<=x2; i++){
            map[y, i] = '─';
        }
        Redraw(x1, y, x2, y);
    }

    public void DrawLineVertical(int y1, int y2, int x){
        for (int i = y1; i<=y2; i++){
            map[i, x] = '│';
        }
        Redraw(x, y1, x, y2);
    }

    private bool HasLeft(int y, int x){
        return !(x==0 || map[y, x-1] == '\0' || map[y, x-1] == '┤' || map[y, x-1] == '│');
    }

    private bool HasRight(int y, int x){
        return !(x==cols || map[y, x+1] == '\0' || map[y, x+1] == '├' || map[y, x+1] == '│');
    }

    private bool HasUp(int y, int x){
        return !(y==0 || map[y-1, x] == '\0' || map[y-1, x] == '┴' || map[y-1, x] == '─');
    }

    private bool HasDown(int y, int x){
        return !(y==rows || map[y+1, x] == '\0' || map[y+1, x] == '┬' || map[y+1, x] == '─');
    }

    private void Redraw (int x1, int y1, int x2, int y2){
        for (int i = y1; i<=y2; i++){
            for (int j = x1; j<=x2; j++){
                if (i==y1 || i==y2 || j==x1 || j==x2){
                    if (HasUp(i, j) && HasDown(i, j) && HasLeft(i, j) && HasRight(i, j)){map[i,j] = '┼';}
                    if (!HasUp(i, j) && HasDown(i, j) && HasLeft(i, j) && HasRight(i, j)){map[i,j] = '┬';}
                    if (HasUp(i, j) && !HasDown(i, j) && HasLeft(i, j) && HasRight(i, j)){map[i,j] = '┴';}
                    if (HasUp(i, j) && HasDown(i, j) && !HasLeft(i, j) && HasRight(i, j)){map[i,j] = '├';}
                    if (HasUp(i, j) && HasDown(i, j) && HasLeft(i, j) && !HasRight(i, j)){map[i,j] = '┤';}
                }
            }
        }
    }

    public char[,] GetMap(){
        return map;
    }
    
}
```

```
public class SymbolSpritesMap(RoadMap rm){
    private char [,] map = rm.GetMap();

    public void PrintSymbolSpriteRoad(char symbol ='█'){
        char [,][,] newMap = new char [map.GetLength(0), map.GetLength(1)][,];
        Dictionary<char, char[,]> sprites = new Dictionary<char, char[,]>{
            {'─', new char[4, 6]
        {
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { ' ', ' ', ' ', ' ', ' ', ' ' }
        }},
        {'│', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'┼', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'┬', new char[4, 6]
        {
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'┴', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { symbol, symbol, symbol, symbol, symbol, symbol },
            { ' ', ' ', ' ', ' ', ' ', ' ' }
        }},
        {'├', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'┤', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'\0', new char[4, 6]
        {
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { ' ', ' ', ' ', ' ', ' ', ' ' }
        }},
        {'┌', new char[4, 6]
        {
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'┐', new char[4, 6]
        {
            { ' ', ' ', ' ', ' ', ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, ' ', ' ' }
        }},
        {'└', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', symbol, symbol, symbol, symbol },
            { ' ', ' ', ' ', ' ', ' ', ' ' }
        }},
        {'┘', new char[4, 6]
        {
            { ' ', ' ', symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { symbol, symbol, symbol, symbol, ' ', ' ' },
            { ' ', ' ', ' ', ' ', ' ', ' ' }
        }},
        };

        for (int i = 0; i<map.GetLength(0); i++){
            for (int j = 0; j<map.GetLength(1); j++){
                char [,] arr = sprites[map[i,j]];
                newMap[i,j] = arr;
            }

            for (int k=0; k<4; k++){
                for (int j = 0; j<map.GetLength(1); j++){
                    for (int l=0; l<6; l++){
                        Console.Write(newMap[i,j][k,l]);
                    }
                }
                Console.WriteLine();
            }
        }
    }
}
```

```
public void PrintTable(char[,] table)
    {
        int rows = table.GetLength(0);
        int columns = table.GetLength(1);

        for (int i = 0; i < rows; i++)
        {
            for (int j = 0; j < columns; j++)
            {
                Console.Write($"{table[i, j]} ");
            }
            Console.WriteLine();
        }
    }
```
